# Processing Sequences Using RNNs and CNNs

# Preparing data
 for machine learning models is a crucial step in the ML pipeline. Below is a complete code example for preparing a dataset for machine learning using Python and common libraries like pandas, scikit-learn, and matplotlib.

This process involves the following steps:

###Loading and exploring the dataset.
###Encoding categorical variables.
###Splitting the data into training and testing sets.
###Visualizing the data.

```python
# Loading and Exploring the Dataset
import pandas as pd
import numpy as np

# Load dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"  # Example URL
columns = ["sepal_length", "sepal_width", "petal_length", "petal_width", "species"]
df = pd.read_csv(url, header=None, names=columns)

# Display first few rows of the dataset
print(df.head())

# Display summary statistics
print(df.describe())

# Check for missing values
print(df.isnull().sum())
```

```python
# Handling missing values (if any)
# For simplicity, we can remove rows with missing values:
df = df.dropna()
print(df.head())

# Alternatively, you can impute missing values with the mean/median/mode for numerical columns:
# df['column_name'] = df['column_name'].fillna(df['column_name'].mean())
```

```python
# Encoding Categorical Variables
from sklearn.preprocessing import LabelEncoder

# Encode the target variable (species)
label_encoder = LabelEncoder()
df['species'] = label_encoder.fit_transform(df['species'])

# Check encoded values
print(df['species'].unique())
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler

# Separate features and target
X = df.drop('species', axis=1)
y = df['species']

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Check scaled data
print(pd.DataFrame(X_scaled, columns=X.columns).head())
```

```python
# Splitting the Data into Training and Testing Sets
from sklearn.model_selection import train_test_split

# Split the data (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Check the shape of training and testing data
print(f"Training features shape: {X_train.shape}")
print(f"Testing features shape: {X_test.shape}")
```

```python
#Visualizing the Data
import matplotlib.pyplot as plt
import seaborn as sns

# Pairplot to visualize relationships between features
sns.pairplot(df, hue='species')
plt.show()

# Correlation matrix to see feature relationships
correlation_matrix = df.drop('species', axis=1).corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

#Forecasting Using a Simple RNN

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split

# Step 1: Load and preprocess the data
# Here we will use a sample dataset. Replace this with your time series dataset.
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
df = pd.read_csv(url, header=0, parse_dates=['Month'], index_col='Month')

# Visualize the dataset
df.plot(figsize=(10, 6))
plt.title('Monthly Airline Passengers')
plt.xlabel('Date')
plt.ylabel('Passengers')
plt.show()

# Step 2: Preprocess the data
# We will scale the data between 0 and 1 using MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df)

# Define the function to create data windows (timesteps)
def create_dataset(data, time_step=1):
    X, y = [], []
    for i in range(len(data) - time_step):
        X.append(data[i:(i + time_step), 0])  # features
        y.append(data[i + time_step, 0])  # target value
    return np.array(X), np.array(y)

# Define the time step (how many past days to look at for predicting future)
time_step = 12

# Create the dataset
X, y = create_dataset(scaled_data, time_step)

# Reshape X to be in the format [samples, time_steps, features] required for RNN
X = X.reshape(X.shape[0], X.shape[1], 1)

# Step 3: Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Step 4: Build the Simple RNN Model
model = Sequential()

# Add SimpleRNN layer
model.add(SimpleRNN(units=50, return_sequences=False, input_shape=(time_step, 1)))

# Add Dense layer for output prediction
model.add(Dense(units=1))

# Compile the model
model.compile(optimizer=Adam(), loss='mean_squared_error')

# Step 5: Train the Model
history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test), verbose=1)

# Step 6: Evaluate the model
# Predict on the test data
predictions = model.predict(X_test)

# Rescale predictions and true values back to the original scale
predictions = scaler.inverse_transform(predictions)
y_test_rescaled = scaler.inverse_transform(y_test.reshape(-1, 1))

# Step 7: Plot the results
plt.figure(figsize=(10, 6))

# Plot true values
plt.plot(y_test_rescaled, color='blue', label='True values')

# Plot predicted values
plt.plot(predictions, color='red', label='Predictions')

plt.title('Time Series Forecasting with Simple RNN')
plt.xlabel('Time')
plt.ylabel('Passengers')
plt.legend()
plt.show()

# Step 8: Calculate RMSE (Root Mean Squared Error)
rmse = np.sqrt(np.mean(np.square(y_test_rescaled - predictions)))
print(f'RMSE: {rmse}')
```

#Forecasting Multivariate Time Series

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.model_selection import train_test_split

# Step 1: Load and Preprocess the Data
# Here we're using a multivariate time series dataset with temperature and humidity data
# You can replace this dataset with your own time series dataset.
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/daily-min-temperatures.csv"
df = pd.read_csv(url, header=0, parse_dates=['Date'], index_col='Date')

# We will add a "humidity" feature for multivariate forecasting. You can replace this with real multivariate data.
df['Humidity'] = np.random.rand(len(df)) * 10  # Just generating random humidity values for example purposes.

# Plot the dataset
df.plot(figsize=(10, 6))
plt.title('Temperature and Humidity Over Time')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend(['Temperature', 'Humidity'])
plt.show()

# Step 2: Preprocess the data
# We need to scale the features so that they are between 0 and 1 for LSTM to work effectively
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df)

# Define the function to create data windows (timesteps) for multivariate time series
def create_dataset(data, time_step=1):
    X, y = [], []
    for i in range(len(data) - time_step):
        X.append(data[i:(i + time_step), :])  # using all features (temperature, humidity)
        y.append(data[i + time_step, 0])  # target is temperature (first column)
    return np.array(X), np.array(y)

# Define the time step (number of previous days to look at to predict future)
time_step = 12

# Create the dataset
X, y = create_dataset(scaled_data, time_step)

# Reshape X to be in the format [samples, time_steps, features]
X = X.reshape(X.shape[0], X.shape[1], X.shape[2])

# Step 3: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Step 4: Build the LSTM model for Multivariate Forecasting
model = Sequential()

# Add LSTM layers
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dropout(0.2))  # Dropout to prevent overfitting

model.add(LSTM(units=50, return_sequences=False))  # Second LSTM layer (no return_sequences)
model.add(Dropout(0.2))

# Add Dense layer for output prediction
model.add(Dense(units=1))  # Predicting the first feature (temperature)

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Step 5: Train the Model
history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test), verbose=1)

# Step 6: Evaluate the model and make predictions
# Predict on the test data
predictions = model.predict(X_test)

# Rescale predictions and true values back to the original scale
predictions_rescaled = scaler.inverse_transform(np.concatenate((predictions, np.zeros_like(predictions)), axis=1))[:, 0]
y_test_rescaled = scaler.inverse_transform(np.concatenate((y_test.reshape(-1, 1), np.zeros_like(y_test.reshape(-1, 1))), axis=1))[:, 0]

# Step 7: Plot the results
plt.figure(figsize=(10, 6))

# Plot true values
plt.plot(y_test_rescaled, color='blue', label='True Values')

# Plot predicted values
plt.plot(predictions_rescaled, color='red', label='Predictions')

plt.title('Multivariate Time Series Forecasting with LSTM')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.legend()
plt.show()

# Step 8: Calculate RMSE (Root Mean Squared Error)
rmse = np.sqrt(np.mean(np.square(y_test_rescaled - predictions_rescaled)))
print(f'RMSE: {rmse}')
```

# Forecasting Several Time Steps Ahead

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.model_selection import train_test_split

# Step 1: Load and preprocess the data
# For this example, we'll use a simple univariate time series dataset like daily temperatures.
# You can replace this with your own multivariate time series dataset if needed.
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/daily-min-temperatures.csv"
df = pd.read_csv(url, header=0, parse_dates=['Date'], index_col='Date')

# Plot the dataset to understand it better
df.plot(figsize=(10, 6))
plt.title('Daily Minimum Temperatures in Melbourne')
plt.xlabel('Date')
plt.ylabel('Temperature (Celsius)')
plt.show()

# Step 2: Preprocess the data
# We will use MinMaxScaler to scale the data between 0 and 1
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df)

# Define the function to prepare the data for multi-step forecasting
def create_dataset(data, time_step=1, forecast_steps=1):
    X, y = [], []
    for i in range(len(data) - time_step - forecast_steps + 1):
        X.append(data[i:(i + time_step), 0])  # Input features (previous time_steps)
        y.append(data[(i + time_step):(i + time_step + forecast_steps), 0])  # Output labels (future forecast_steps)
    return np.array(X), np.array(y)

# Define time_step (number of previous days to use) and forecast_steps (how many future days to predict)
time_step = 12  # Look at the past 12 days
forecast_steps = 7  # Predict the next 7 days

# Prepare the dataset
X, y = create_dataset(scaled_data, time_step, forecast_steps)

# Reshape X to be in the format [samples, time_steps, features] for LSTM
X = X.reshape(X.shape[0], X.shape[1], 1)

# Step 3: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Step 4: Build the LSTM model for multi-step forecasting
model = Sequential()

# Add LSTM layers
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(Dropout(0.2))  # Dropout to prevent overfitting

model.add(LSTM(units=50, return_sequences=False))  # Second LSTM layer (no return_sequences)
model.add(Dropout(0.2))

# Add Dense layer for multi-step output
model.add(Dense(units=forecast_steps))  # Predicting multiple steps ahead

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Step 5: Train the Model
history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test), verbose=1)

# Step 6: Evaluate the model and make multi-step predictions
predictions = model.predict(X_test)

# Rescale predictions and true values back to the original scale
predictions_rescaled = scaler.inverse_transform(np.concatenate((predictions, np.zeros_like(predictions)), axis=1))[:, 0]
y_test_rescaled = scaler.inverse_transform(np.concatenate((y_test, np.zeros_like(y_test)), axis=1))[:, 0]

# Step 7: Plot the results
plt.figure(figsize=(10, 6))

# Plot true values (multi-step)
for i in range(len(y_test_rescaled) - forecast_steps + 1):  # Ensure we don't go out of bounds
    plt.plot(range(i, i + forecast_steps), y_test_rescaled[i:i+forecast_steps], color='blue', alpha=0.5)

# Plot predicted values (multi-step)
for i in range(len(predictions_rescaled) - forecast_steps + 1):  # Ensure we don't go out of bounds
    plt.plot(range(i, i + forecast_steps), predictions_rescaled[i:i+forecast_steps], color='red', alpha=0.5)

plt.title('Multi-Step Time Series Forecasting with LSTM')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.show()

# Step 8: Calculate RMSE (Root Mean Squared Error) for multi-step predictions
rmse = np.sqrt(np.mean(np.square(y_test_rescaled[:len(predictions_rescaled)] - predictions_rescaled)))
print(f'RMSE: {rmse}')
```

# forecasting using a Sequence-to-Sequence model

```python
# forecasting using a Sequence-to-Sequence model
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, RepeatVector, TimeDistributed
from sklearn.model_selection import train_test_split

# Step 1: Load and preprocess the data
# For this example, we'll use the Daily Minimum Temperatures dataset.
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/daily-min-temperatures.csv"
df = pd.read_csv(url, header=0, parse_dates=['Date'], index_col='Date')

# Plot the dataset to understand it better
df.plot(figsize=(10, 6))
plt.title('Daily Minimum Temperatures in Melbourne')
plt.xlabel('Date')
plt.ylabel('Temperature (Celsius)')
plt.show()

# Step 2: Preprocess the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df)

# Define the function to prepare the data for Seq2Seq forecasting
def create_dataset(data, time_step=1, forecast_steps=1):
    X, y = [], []
    for i in range(len(data) - time_step - forecast_steps + 1):
        X.append(data[i:(i + time_step), 0])  # Input sequence (past time_steps)
        y.append(data[(i + time_step):(i + time_step + forecast_steps), 0])  # Output sequence (future forecast_steps)
    return np.array(X), np.array(y)

# Define time_step (number of previous days to use) and forecast_steps (how many future days to predict)
time_step = 12  # Use the past 12 days of data
forecast_steps = 7  # Forecast the next 7 days

# Prepare the dataset
X, y = create_dataset(scaled_data, time_step, forecast_steps)

# Reshape X to be in the format [samples, time_steps, features] for LSTM
X = X.reshape(X.shape[0], X.shape[1], 1)

# Step 3: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Step 4: Build the Seq2Seq model
model = Sequential()

# Encoder: LSTM layer
model.add(LSTM(units=50, activation='relu', input_shape=(X_train.shape[1], 1), return_sequences=False))

# Repeat the encoded vector for each forecast step (necessary for the decoder)
model.add(RepeatVector(forecast_steps))

# Decoder: LSTM layer
model.add(LSTM(units=50, activation='relu', return_sequences=True))

# TimeDistributed layer to apply Dense layer to each time step in the sequence
model.add(TimeDistributed(Dense(1)))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Step 5: Train the model
history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test), verbose=1)

# Step 6: Make multi-step predictions
predictions = model.predict(X_test)

# Rescale predictions and true values back to the original scale
predictions_rescaled = scaler.inverse_transform(np.concatenate((predictions.reshape(predictions.shape[0], predictions.shape[1]), np.zeros_like(predictions.reshape(predictions.shape[0], predictions.shape[1]))), axis=1))[:, 0]
y_test_rescaled = scaler.inverse_transform(np.concatenate((y_test, np.zeros_like(y_test)), axis=1))[:, 0]

# Step 7: Plot the results
plt.figure(figsize=(10, 6))

# Plot true values (multi-step)
for i in range(len(y_test_rescaled) - forecast_steps + 1):  # Make sure there are enough data points for each forecast step
    plt.plot(range(i, i + forecast_steps), y_test_rescaled[i:i+forecast_steps], color='blue', alpha=0.5)

# Plot predicted values (multi-step)
for i in range(len(predictions_rescaled) - forecast_steps + 1):  # Make sure there are enough data points for each forecast step
    plt.plot(range(i, i + forecast_steps), predictions_rescaled[i:i+forecast_steps], color='red', alpha=0.5)

plt.title('Seq2Seq Time Series Forecasting with LSTM')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.show()


```

# Fighting Unstable Gradients with LSTM

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split

# Step 1: Load and preprocess the data
# Using the Daily Minimum Temperatures dataset for demonstration
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/daily-min-temperatures.csv"
df = pd.read_csv(url, header=0, parse_dates=['Date'], index_col='Date')

# Plot the dataset to understand it better
df.plot(figsize=(10, 6))
plt.title('Daily Minimum Temperatures in Melbourne')
plt.xlabel('Date')
plt.ylabel('Temperature (Celsius)')
plt.show()

# Step 2: Preprocess the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df)

# Function to create dataset for LSTM (like the previous example)
def create_dataset(data, time_step=1):
    X, y = [], []
    for i in range(len(data) - time_step):
        X.append(data[i:(i + time_step), 0])
        y.append(data[i + time_step, 0])
    return np.array(X), np.array(y)

# Define time_step (number of previous days to use)
time_step = 12  # Using the past 12 days of data

# Prepare the dataset
X, y = create_dataset(scaled_data, time_step)

# Reshape X to be in the format [samples, time_steps, features] for LSTM
X = X.reshape(X.shape[0], X.shape[1], 1)

# Step 3: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Step 4: Build the LSTM model
model = Sequential()

# Adding LSTM layer with 50 units and input shape
model.add(LSTM(units=50, activation='relu', input_shape=(X_train.shape[1], 1)))

# Add a Dense layer for the output
model.add(Dense(units=1))

# Step 5: Compile the model with gradient clipping
optimizer = Adam(clipvalue=1.0)  # Clip gradients to a maximum value of 1.0 to avoid exploding gradients

model.compile(optimizer=optimizer, loss='mean_squared_error')

# Step 6: Train the model
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test), verbose=1)

# Step 7: Make predictions
predictions = model.predict(X_test)

# Rescale predictions and true values back to the original scale
predictions_rescaled = scaler.inverse_transform(np.concatenate((predictions, np.zeros_like(predictions)), axis=1))[:, 0]
y_test_rescaled = scaler.inverse_transform(np.concatenate((y_test.reshape(-1, 1), np.zeros_like(y_test.reshape(-1, 1))), axis=1))[:, 0]

# Step 8: Plot the results
plt.figure(figsize=(10, 6))
plt.plot(y_test_rescaled, color='blue', label='True Values')
plt.plot(predictions_rescaled, color='red', label='Predicted Values')
plt.title('Temperature Forecasting with LSTM (Gradient Clipping)')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.legend()
plt.show()

# Step 9: Calculate RMSE (Root Mean Squared Error)
rmse = np.sqrt(np.mean(np.square(y_test_rescaled - predictions_rescaled)))
print(f'RMSE: {rmse}')
```

# Using 1D convolutional layers to process sequences

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split

# Step 1: Load and preprocess the data
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/daily-min-temperatures.csv"
df = pd.read_csv(url, header=0, parse_dates=['Date'], index_col='Date')

# Plot the dataset to understand it better
df.plot(figsize=(10, 6))
plt.title('Daily Minimum Temperatures in Melbourne')
plt.xlabel('Date')
plt.ylabel('Temperature (Celsius)')
plt.show()

# Step 2: Preprocess the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df)

# Function to create dataset for Conv1D (like previous examples)
def create_dataset(data, time_step=1):
    X, y = [], []
    for i in range(len(data) - time_step):
        X.append(data[i:(i + time_step), 0])
        y.append(data[i + time_step, 0])
    return np.array(X), np.array(y)

# Define time_step (number of previous days to use)
time_step = 12  # Using the past 12 days of data

# Prepare the dataset
X, y = create_dataset(scaled_data, time_step)

# Reshape X to be in the format [samples, time_steps, features] for Conv1D
X = X.reshape(X.shape[0], X.shape[1], 1)

# Step 3: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Step 4: Build the model using Conv1D layers
model = Sequential()

# Adding 1D convolutional layer with 64 filters, kernel size 3, and activation function 'relu'
model.add(Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(X_train.shape[1], 1)))

# Adding a max pooling layer to reduce the dimensionality
model.add(MaxPooling1D(pool_size=2))

# Adding another 1D convolutional layer
model.add(Conv1D(filters=32, kernel_size=3, activation='relu'))

# Adding another max pooling layer
model.add(MaxPooling1D(pool_size=2))

# Flattening the output of the convolutional layers to feed into the dense layers
model.add(Flatten())

# Adding a fully connected layer
model.add(Dense(units=50, activation='relu'))

# Output layer (single neuron for regression)
model.add(Dense(units=1))

# Step 5: Compile the model
model.compile(optimizer=Adam(), loss='mean_squared_error')

# Step 6: Train the model
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test), verbose=1)

# Step 7: Make predictions
predictions = model.predict(X_test)

# Rescale predictions and true values back to the original scale
predictions_rescaled = scaler.inverse_transform(np.concatenate((predictions, np.zeros_like(predictions)), axis=1))[:, 0]
y_test_rescaled = scaler.inverse_transform(np.concatenate((y_test.reshape(-1, 1), np.zeros_like(y_test.reshape(-1, 1))), axis=1))[:, 0]

# Step 8: Plot the results
plt.figure(figsize=(10, 6))
plt.plot(y_test_rescaled, color='blue', label='True Values')
plt.plot(predictions_rescaled, color='red', label='Predicted Values')
plt.title('Temperature Forecasting using 1D Convolutional Neural Network')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.legend()
plt.show()

# Step 9: Calculate RMSE (Root Mean Squared Error)
rmse = np.sqrt(np.mean(np.square(y_test_rescaled - predictions_rescaled)))
print(f'RMSE: {rmse}')
```
